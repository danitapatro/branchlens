```python
import pandas as pd
import numpy as np
import random
 
np.random.seed(42)
random.seed(42)
 
BRANCHES = [
    ("BRN001","Anna Nagar","Chennai","South"),
    ("BRN002","T Nagar","Chennai","South"),
    ("BRN003","Velachery","Chennai","South"),
    ("BRN004","Adyar","Chennai","South"),
    ("BRN005","Tambaram","Chennai","South"),
    ("BRN006","Koramangala","Bangalore","South"),
    ("BRN007","Indiranagar","Bangalore","South"),
    ("BRN008","Whitefield","Bangalore","South"),
    ("BRN009","Jayanagar","Bangalore","South"),
    ("BRN010","Hebbal","Bangalore","South"),
    ("BRN011","Banjara Hills","Hyderabad","South"),
    ("BRN012","Hitech City","Hyderabad","South"),
    ("BRN013","Secunderabad","Hyderabad","South"),
    ("BRN014","Kukatpally","Hyderabad","South"),
    ("BRN015","Connaught Place","Delhi","North"),
    ("BRN016","Dwarka","Delhi","North"),
    ("BRN017","Lajpat Nagar","Delhi","North"),
    ("BRN018","Rohini","Delhi","North"),
    ("BRN019","Gurgaon Sec 29","Gurgaon","North"),
    ("BRN020","Gurgaon Sec 14","Gurgaon","North"),
    ("BRN021","Noida Sec 18","Noida","North"),
    ("BRN022","Noida Sec 62","Noida","North"),
    ("BRN023","Chandigarh Sec17","Chandigarh","North"),
    ("BRN024","Andheri West","Mumbai","West"),
    ("BRN025","Bandra","Mumbai","West"),
    ("BRN026","Lower Parel","Mumbai","West"),
    ("BRN027","Borivali","Mumbai","West"),
    ("BRN028","Thane West","Thane","West"),
    ("BRN029","Viman Nagar","Pune","West"),
    ("BRN030","Kothrud","Pune","West"),
    ("BRN031","Hinjewadi","Pune","West"),
    ("BRN032","CG Road","Ahmedabad","West"),
    ("BRN033","Navrangpura","Ahmedabad","West"),
    ("BRN034","Rajkot Main","Rajkot","West"),
    ("BRN035","Kankurgachi","Kolkata","East"),
    ("BRN036","Salt Lake","Kolkata","East"),
    ("BRN037","Park Street","Kolkata","East"),
    ("BRN038","Howrah","Kolkata","East"),
    ("BRN039","Bhubaneswar CBD","Bhubaneswar","East"),
    ("BRN040","Cuttack Main","Cuttack","East"),
    ("BRN041","Patna Main","Patna","East"),
    ("BRN042","Ranchi Main","Ranchi","East"),
    ("BRN043","Guwahati GS Road","Guwahati","East"),
    ("BRN044","Jaipur MI Road","Jaipur","North"),
    ("BRN045","Jaipur Vaishali","Jaipur","North"),
    ("BRN046","Lucknow Hazratganj","Lucknow","North"),
    ("BRN047","Kanpur Main","Kanpur","North"),
    ("BRN048","Kochi MG Road","Kochi","South"),
    ("BRN049","Coimbatore RS Puram","Coimbatore","South"),
    ("BRN050","Madurai Main","Madurai","South"),
]
 
TIERS = ["high"]*16 + ["mid"]*22 + ["low"]*12
random.shuffle(TIERS)
TIER_MAP = {b[0]: TIERS[i] for i, b in enumerate(BRANCHES)}
 
MONTHS = pd.date_range("2021-01-01", periods=36, freq="MS")
 
BRANCH_TYPE_MAP = {}
for b in BRANCHES:
    city = b[2]
    if city in ["Mumbai","Delhi","Bangalore","Chennai"]:
        BRANCH_TYPE_MAP[b[0]] = "Metro"
    elif city in ["Hyderabad","Pune","Kolkata","Ahmedabad"]:
        BRANCH_TYPE_MAP[b[0]] = "Urban"
    elif city in ["Noida","Gurgaon","Thane","Chandigarh","Kochi","Coimbatore","Jaipur","Lucknow"]:
        BRANCH_TYPE_MAP[b[0]] = "Semi-Urban"
    else:
        BRANCH_TYPE_MAP[b[0]] = "Rural"
 
SEASONAL = {1:0.87,2:0.90,3:0.96,4:0.98,5:1.00,6:1.02,
            7:1.03,8:1.05,9:1.04,10:1.07,11:1.11,12:1.16}
YEAR_GROWTH = {2021:1.00, 2022:1.09, 2023:1.18}
 
rows = []
 
for branch_id, branch_name, city, region in BRANCHES:
    tier  = TIER_MAP[branch_id]
    btype = BRANCH_TYPE_MAP[branch_id]
 
    if tier == "high":
        base_txn     = np.random.randint(4000, 6000)
        base_rev     = np.random.uniform(30, 50)
        base_exp     = np.random.uniform(16, 25)
        base_loan    = np.random.uniform(200, 360)
        base_dep     = np.random.uniform(250, 450)
        base_csat    = np.random.uniform(3.9, 4.7)
        headcount    = np.random.randint(20, 32)
        new_acc_base = np.random.randint(180, 320)
    elif tier == "mid":
        base_txn     = np.random.randint(2000, 4000)
        base_rev     = np.random.uniform(15, 30)
        base_exp     = np.random.uniform(10, 19)
        base_loan    = np.random.uniform(80, 200)
        base_dep     = np.random.uniform(100, 250)
        base_csat    = np.random.uniform(3.3, 4.1)
        headcount    = np.random.randint(11, 20)
        new_acc_base = np.random.randint(80, 180)
    else:
        base_txn     = np.random.randint(700, 2000)
        base_rev     = np.random.uniform(5, 15)
        base_exp     = np.random.uniform(7, 13)
        base_loan    = np.random.uniform(25, 80)
        base_dep     = np.random.uniform(35, 100)
        base_csat    = np.random.uniform(2.4, 3.5)
        headcount    = np.random.randint(4, 11)
        new_acc_base = np.random.randint(20, 80)
 
    for month in MONTHS:
        season    = SEASONAL[month.month]
        yr_growth = YEAR_GROWTH[month.year]
        noise     = np.random.uniform(0.91, 1.09)
        factor    = season * yr_growth * noise
 
        transactions    = int(base_txn * factor)
        revenue         = round(base_rev * factor, 2)
        expenses        = round(base_exp * np.random.uniform(0.94, 1.06), 2)
        loan_disbursed  = round(base_loan * factor, 2)
        deposits        = round(base_dep * factor, 2)
        csat            = round(np.clip(base_csat + np.random.uniform(-0.35, 0.35), 1.0, 5.0), 2)
        new_accounts    = int(new_acc_base * factor)
        npa_ratio       = round(np.random.uniform(0.5, 5.5) if tier == "low" else
                                np.random.uniform(0.2, 3.0) if tier == "mid" else
                                np.random.uniform(0.1, 1.8), 2)
        digital_txn_pct = round(np.random.uniform(30, 55) if btype == "Rural" else
                                 np.random.uniform(50, 75) if btype == "Semi-Urban" else
                                 np.random.uniform(65, 88), 2)
 
        profit            = round(revenue - expenses, 2)
        profit_margin     = round((profit / revenue) * 100, 2)
        rev_per_txn       = round((revenue * 100000) / transactions, 2)
        cost_efficiency   = round(expenses / revenue, 4)
        revenue_per_staff = round(revenue / headcount, 4)
 
        rows.append({
            "branch_id":              branch_id,
            "branch_name":            branch_name,
            "city":                   city,
            "region":                 region,
            "branch_type":            btype,
            "tier":                   tier,
            "month":                  month.strftime("%Y-%m"),
            "year":                   month.year,
            "month_num":              month.month,
            "quarter":                f"Q{(month.month-1)//3+1}",
            "headcount":              headcount,
            "total_transactions":     transactions,
            "new_accounts_opened":    new_accounts,
            "revenue_lakhs":          revenue,
            "expenses_lakhs":         expenses,
            "profit_lakhs":           profit,
            "profit_margin_pct":      profit_margin,
            "loan_disbursed_lakhs":   loan_disbursed,
            "deposits_lakhs":         deposits,
            "npa_ratio_pct":          npa_ratio,
            "digital_txn_pct":        digital_txn_pct,
            "revenue_per_txn_inr":    rev_per_txn,
            "cost_efficiency":        cost_efficiency,
            "revenue_per_staff":      revenue_per_staff,
            "customer_satisfaction":  csat,
        })
 
df = pd.DataFrame(rows)
 
# --- Anomaly 1: BRN035 Kankurgachi — CSAT crash + txn drop in Q3 2022 ---
m1 = (df.branch_id=="BRN035") & (df.year==2022) & (df.month_num.isin([7,8,9]))
df.loc[m1,"customer_satisfaction"] = (df.loc[m1,"customer_satisfaction"] - 1.5).clip(lower=1.0).round(2)
df.loc[m1,"total_transactions"]    = (df.loc[m1,"total_transactions"] * 0.72).astype(int)
 
# --- Anomaly 2: BRN026 Lower Parel — expense spike Nov 2023 (renovation) ---
m2 = (df.branch_id=="BRN026") & (df.year==2023) & (df.month_num==11)
df.loc[m2,"expenses_lakhs"]    = (df.loc[m2,"expenses_lakhs"] * 1.70).round(2)
df.loc[m2,"profit_lakhs"]      = (df.loc[m2,"revenue_lakhs"] - df.loc[m2,"expenses_lakhs"]).round(2)
df.loc[m2,"profit_margin_pct"] = ((df.loc[m2,"profit_lakhs"] / df.loc[m2,"revenue_lakhs"]) * 100).round(2)
df.loc[m2,"cost_efficiency"]   = (df.loc[m2,"expenses_lakhs"] / df.loc[m2,"revenue_lakhs"]).round(4)
 
# --- Anomaly 3: BRN047 Kanpur — persistently high NPA (stressed branch) ---
m3 = df.branch_id=="BRN047"
df.loc[m3,"npa_ratio_pct"] = (df.loc[m3,"npa_ratio_pct"] + 4.5).clip(upper=9.9).round(2)
 
# --- Anomaly 4: BRN010 Hebbal — digital adoption spike in 2023 ---
m4 = (df.branch_id=="BRN010") & (df.year==2023)
df.loc[m4,"digital_txn_pct"] = (df.loc[m4,"digital_txn_pct"] + 15).clip(upper=98).round(2)
 
# --- Save ---
df.to_csv(r"C:\Users\hp\Desktop\Data_Analysis_projects\BranchLens\branch_data.csv", index=False)
 
print(f"Total rows     : {len(df):,}")
print(f"Total columns  : {len(df.columns)}")
print(f"Branches       : {df['branch_id'].nunique()}")
print(f"Years covered  : {sorted(df['year'].unique())}")
print(f"\nColumns: {list(df.columns)}")
print(f"\nKey stats:\n{df[['revenue_lakhs','profit_margin_pct','npa_ratio_pct','customer_satisfaction']].describe().round(2)}")
 
```
